Using the data from the BRFSS, compute the linear least squares fit for log(weight) versus height. How would you best present the estimated parameters for a model like this where one of the variables is log-transformed? If you were trying to guess someoneâ€™s weight, how much would it help to know their height? 

```python
import brfss as bs
import pandas, math, thinkstats2
import matplotlib.pyplot as plt
df = bs.ReadBrfss()
df.dropna(inplace=True)

df['logWgt'] = [math.log(val) for val in df.wtkg2]
plt.scatter(df.logWgt, df.htm3, alpha=.1)
```

```python
inter, slope = thinkstats2.LeastSquares(df.logWgt, df.htm3)

fit_x, fit_y = thinkstats2.FitLine(df.logWgt, inter, slope)
print df.logWgt.mean(), math.exp(df.logWgt.mean())
print df.htm3.mean(), slope
print df.htm3.std()
plt.scatter(df.logWgt, df.htm3, alpha=.1)
plt.plot(fit_x, fit_y)

```

```python
thinkstats2.IQR(df.htm3)
```

The average height is 169 centimeters, and for every change of weight by a factor of 10, the height will change by roughly 23 centimeters. The IQR is 15, so knowing the height of someone allows you to guess their weight within 7.5 pounds with 50% confidence.
